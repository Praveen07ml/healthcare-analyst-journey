```python
# Check if "A man, a plan, a canal: Panama" is a palindrome.
# Ignore spaces, punctuation, and case.


words = "A man, a plan, a canal: Panama"

# so we need to clean it remove spaces,commas etc

clean_data = []

for char in words:
    if char.isalnum():
        clean_data.append(char.lower())

print(clean_data)

palindrome = True

for i in range(len(clean_data)//2):
    if clean_data[i] != clean_data[-1-i]:
        palindrome = False
        break

print()

if palindrome:
    print("Its palindrome")
else:
    print("Not Palindrome")

        
```

```python
# Reverse only the VOWELS in "hello" to get "holle"

word = list("hello")

vowels = 'aeiou'

left = 0
right = len(word) -1

while left < right:
    
    if word[left] not in vowels:
        left += 1
    elif word[right] not in vowles:
        right -= 1
    else:
        word[left],word[right] = word[right],word[left]
        left += 1
        right -= 1

print("".join(word))
    

        
        
```

```python
# Remove adjacent duplicate characters: "abbaca" → "ca"

# its taken so much time for me to understand its eventually stack problem but i understand logic 

word = "abbaca"

stack = []

for char in word:
    if stack and stack[-1] == char:
        stack.pop()

    else:
        stack.append(char)

print("".join(stack))
```

```python
# Compress string "aabcccccaaa" to "a2b1c5a3"


word = "aabcccccaaa"
result = []

count = 1

for i in range(1,len(word)):
    if word[i] == word[i - 1]:
        count += 1
    else:
        result.append(word[i-1] + str(count))

result.append(word[i-1] + str(count))

print("".join(result))

```
