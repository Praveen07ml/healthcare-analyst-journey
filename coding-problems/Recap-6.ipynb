```python
#reverse a list without using reverse 

numbers = [1,2,3,4,5]
result = []
for i in range(len(numbers)):
    result.append(numbers[-1-i])
print(result)
    
```

```python
# check if a list is palindrome

numbers = [1,2,3,4,5]
valid = True
for i in range(len(numbers)//2):
    if numbers[i] != numbers[-1-i]:
        valid = False
        break
if valid:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# find pair of numbers in [2,7,11,15] up to 9

numbers = [2,7,11,15]
total = 9
result = []

for i in range(len(numbers)):
    for j in range(i+1,len(numbers)):
        value = numbers[i] + numbers[j]
        if value  == total:
            result.append([numbers[i],numbers[j]])
print(result)
        
```

```python
# remove duplicates from sorted list

numbers = [1,1,2,2,3,4,4,5]
result = []
for i in range(len(numbers) - 1):
    if numbers[i] != numbers[i+1]:
        result.append(numbers[i])
print(result)
```

```python
 # remove duplicates from sorted list

numbers = [1,1,2,2,3,4,4,5]
unique = []
for number in numbers:
    if number not in unique:
        unique.append(number)
print(unique)

# even thought its not sorted array this way its still works
```

```python
# reverse a string without using slicing

word = "Hello"
reverse = []
for i in range(len(word)):
    reverse.append(word[-1-i])
    
print("".join(reverse))
```

```python
# check the two strings are anagrams

word1 = "listen"
word2 = "silent"

word1 = sorted(word1)
word2 = sorted(word2)
valid = True

for i in range(len(word1)):
    if word1[i] != word2[i]:
        valid = False
        break
if valid:
    print("Anagrams")
else:
    print("Not Anagrams")
```
