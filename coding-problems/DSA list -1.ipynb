- Today List manipulation MiN/MAX pattern

```python
# Find the largest number in [34, 12, 78, 23, 90, 45, 67]

numbers = [34, 12, 78, 23, 90, 45, 67]
largest = numbers[0]

for number in numbers:
    if number > largest:
        largest = number
print(largest)
```

```python
# Find the SMALLEST number in [45, 12, 8, 67, 23, 5, 34]

numbers = [45, 12, 8, 67, 23, 5, 34]
smallest = numbers[0]

for number in numbers:
    if number < smallest:
        smallest = number
print(smallest)
```

```python
# Find the largest EVEN number in [15, 22, 8, 45, 18, 33, 7]

numbers = [15, 22, 8, 45, 18, 33, 7]
largest_even = numbers[0]

for number in numbers:
    if number % 2 == 0 and number > largest_even:
        largest_even = number
print(largest_even)
```

```python
# Find BOTH the maximum AND minimum in one loop for [34, 12, 78, 23, 90, 45, 67]

numbers = [34, 12, 78, 23, 90, 45, 67]
largest = numbers[0]
smallest = numbers[0]

for number in numbers:
    if number > largest:
        largest = number
    if number < smallest:
        smallest = number
print(largest)
print(smallest)
```
