```python
numbers = list(range(1,21))
result = []
for number in numbers:
    if number > 1:
        prime = True
        for i in range(2,number):
            if number % i == 0:
                prime = False
                break
        if prime:
            result.append(number)
print(result)
```

```python
word = "madam"
left = 0
right = len(word) -1
result = True
while left < right:
    if word[left] != word[right]:
        result = False
        break
    left += 1
    right -= 1
print(result)
```
