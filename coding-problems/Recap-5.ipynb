```python
# word statistics
# paragraph = "python is great.python is powerfull learning python is fun"
# total words
# most common word
#Average words length
# HOw many unique words

paragraph = "python is great.python is powerfull learning python is fun"

# cleaning paragraph 

paragraph = paragraph.replace("."," ")
words = paragraph.split()

# finding words count 

freq = {}
total_words = len(words)
for word in words:
    freq[word] = freq.get(word,0) + 1

print("words frequency - ", freq)
print("total words - ", total_words)

max_count = max(freq.values())
common_words = [key for key,value in freq.items() if value == max_count]
print("Common Words :",common_words)

Average_words = sum( len(word) for word in words) / total_words
print("Average Words Length - ", Average_words)

unique_words = [key for key,value in freq.items() if value == 1]
print("Unique_words - ", unique_words)

```

```python
# Shopping kart total

items = [('apple',50,3),('banana',30,5),('orange',40,2)]

# total cost
# most expensive item
# which items cost more than 100

freq = {}
for item in items:
    freq[item[0]] = item[1] * item[2]
    
        

print("Kart_Freq : ", freq)

total_cost = sum(freq.values())
print("Total Cost :",total_cost)


expensive = {}
for item in items:
    expensive[item[0]] = item[1]
max_value = max(expensive.values())
most_expensive = [key for key in expensive if expensive[key] == max_value]
print("Most Expensive Item : ",most_expensive)

more_expensive = [key for key,value in freq.items() if value > 100]
print("More than 100 items : ", more_expensive)


```

```python
# check if string "racecar" is palindrome

word = "racecar"
left = 0
right = len(word) - 1
result = True
while left < right:
    if word[left] != word[right]:
        result = False
        break
    left += 1
    right -= 1
print(result)
```

```python
# find the second largest number in the given list

numbers = [45,78,23,90,67,34,88]

first = numbers[0]
second = numbers[1]

for number in numbers:
    if number > first:
        second = first
        first = number
    if number > second and number != first:
        second = number
print(first)
print(second)
```

```python
# count frequency of each digit in  "123112233"

numbers = "123112233"

freq = {}
for number in numbers:
    if number in freq:
        freq[number] = freq.get(number,0) + 1
    else:
        freq[number] = 1
print(freq)
```

```python
# count how many numbers from 1 to 50 are perfect cubes

numbers = list(range(1,51))
count = 0
for number in numbers:
    if int(number ** 0.34)**3 == number:
        count += 1
print(count)
```

```python
# count how many times "a" appear in each word

words = "Banana apple papaya"
count = {}
for word in words.split():
   for char in word:
       if char == 'a':
           if word in count:
               count[word] += 1
           else:
               count[word] = 1
print(count)

```

```python
# sum of sqaures of all prime numbers from 1 to 20

total_squares = 0
for number in range(1,21):
    if number > 1:
        prime = True
        for i in range(2,number):
            if number % i == 0:
                prime = False
                break
        if prime:
            total_squares += (number ** 2)
print(total_squares)
```
