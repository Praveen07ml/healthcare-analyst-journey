```python
word = "Hello"
freq = {}

for char in word:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1
print(freq)
```

```python
word = "programming"
freq = {}
for char in word:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1
print(freq)

max_value = max(freq.values())

for key in freq.keys():
    if freq[key] == max_value:
        print(f"{key} : {freq[key]}")
```

```python
word = "banana"
freq = {}

for char in word:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)

for key,value in freq.items():
    if value > 1:
        print(f"{key}:{value}")
```

```python
total = 0

for number in range(1,31):
    if number > 1:
        prime = True
        for i in range(2,number):
            if number % i == 0:
                prime = False
                break
        if prime:
            total += number
print(total)
```
