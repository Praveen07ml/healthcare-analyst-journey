# Two Pointers (Fast/Slow) arrangement pettern 

```python
# Move all zeros to the end: [0, 1, 0, 3, 12] → [1, 3, 12, 0, 0]

numbers = [0, 1, 0, 3, 12]

slow = 0

for fast in range(len(numbers)):
    if numbers[fast] != 0:
        numbers[slow] = numbers[fast]
        slow += 1

# for remaining you need to assign zeroes to end

for i in range(slow,len(numbers)):
    numbers[i] = 0

print(numbers)
```

```python
# Remove all occurrences of value 3 from [3, 2, 2, 3, 4] to get [2, 2, 4]

numbers = [3, 2, 2, 3, 4]

slow = 0

for fast in range(len(numbers)):
    if numbers[fast] != 3:
        numbers[slow] = numbers[fast]
        slow += 1

print(numbers[:3])


# Alternative


numbers1 = [3, 2, 2, 3, 4]

result = []

for number in numbers1:
    if number != 3:
        result.append(number)

print(result)
```

```python
# Remove duplicates from SORTED array [1, 1, 2, 2, 3, 4, 4, 5]

numbers = [1, 1, 2, 2, 3, 4, 4, 5]

slow = 1

for fast in range(1,len(numbers)):
    if numbers[fast] != numbers[fast - 1]:
        numbers[slow] = numbers[fast]
        slow += 1

print(numbers[:slow])
```

```python
# Move all EVEN numbers to front, ODD to back: [1, 2, 3, 4, 5, 6]

numbers = [1, 2, 3, 4, 5, 6]

slow = 0

for fast in range(len(numbers)):
    if numbers[fast] % 2 == 0:
        numbers[slow],numbers[fast] = numbers[fast],numbers[slow]
        slow += 1


print(numbers)
```
