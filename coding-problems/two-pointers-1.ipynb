# Two Pointers ( Opposite Direction ) pattern

```python
#  Check if list [1, 2, 3, 2, 1] is a palindrome (reads same forwards and backwards)

numbers = [1, 2, 3, 2, 1]

result = True

left = 0
right = len(numbers) - 1

while left < right :

    if numbers[left] != numbers[right]:
        result = False
        break

    left += 1
    right -= 1

if result:
    print("It is Palindrome")
else:
    print("Not Palindrome")

```

```python
# Check if string "racecar" is a palindrome

word = "racecar"

result = True

# Another approach

for i in range(len(numbers)//2):
    if word[i] != word[-1-i]:
        result = False
        break

if result:
    print("It is palindrome")
else:
    print("Its not palindrome")
```

```python
# In SORTED array [1, 2, 3, 4, 6], find TWO numbers that sum to 6.

numbers = [1, 2, 3, 4, 6]
result = []

left = 0
right = len(numbers) -1

target = 6

while left < right:
    
    total = numbers[left] + numbers[right]
    if total == target:
        result.append([numbers[left],numbers[right]])
        left += 1
        right -= 1
        
    elif total < target:
        left += 1

    else:
        right -= 1
    
print(result)
```

```python
# Reverse list [1, 2, 3, 4, 5] IN-PLACE using two-pointer swap

numbers = [1, 2, 3, 4, 5]

left = 0
right = len(numbers) - 1

while left < right:

    numbers[left],numbers[right] = numbers[right],numbers[left]
    left += 1
    right -= 1

print(numbers)
```
