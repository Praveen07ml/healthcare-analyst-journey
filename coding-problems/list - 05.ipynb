## Lists combining / Merging Patttern

```python
# Merge two SORTED lists [1, 3, 5] and [2, 4, 6] into [1, 2, 3, 4, 5, 6]

a = [1, 3, 5]
b = [2, 4, 6]

i = 0
j = 0

result = []

while i < len(a) and j < len(b):
    if a[i] < b[j]:
        result.append(a[i])
        i += 1
    else:
        result.append(b[j])
        j += 1

# for remaining

while i < len(a):
    result.append(a[i])
    i += 1

while j < len(b):
    result.append(b[j])
    j+= 1

print(result)
```

```python
# Merge [1, 4, 7] and [2, 3, 5, 8, 9] into one sorted list

list1 = [1, 4, 7]
list2 = [2, 3, 5, 8, 9]

result = []

i = 0
j = 0

while i < len(list1) and  j < len(list2):
    if list1[i] < list2[j]:
        result.append(list1[i])
        i += 1
    else:
        result.append(list2[j])
        j += 1

# for remaining 

while i < len(list1):
    result.append(list1[i])
    i += 1

while j < len(list2):
    result.append(list2[j])
    j += 1

print(result)
```

```python
# Find COMMON elements in [1, 2, 3, 4, 5] and [3, 4, 5, 6, 7]

list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]


set2 = set(list2)
result = []




for number in list1:
    if number in set2:
        result.append(number)
print(result)
    
```

```python
# Combine [1, 2, 3] and [4, 5, 6] by alternating elements: [1, 4, 2, 5, 3, 6]

list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = []

i = 0
j = 0

while i < len(list1) and j < len(list2):
    result.append(list1[i])
    result.append(list2[j])
    i += 1
    j += 1

# for remaining

while i < len(list1):
    result.append(list1[i])
    i += 1

while j < len(list2):
    result.append(list2[j])
    j += 1

print(result)
    
    
```
