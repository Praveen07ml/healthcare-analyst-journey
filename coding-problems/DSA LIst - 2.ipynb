## List Reversal patterns

```python
# Reverse [1, 2, 3, 4, 5] to get [5, 4, 3, 2, 1]
# Do NOT use .reverse() or [::-1]

# two pointer approach

numbers = [1, 2, 3, 4, 5]
left = 0
right = len(numbers) -1

while left < right:
    numbers[left],numbers[right] = numbers[right],numbers[left]
    left += 1
    right -= 1
    
print(numbers)
```

```python
# Reverse the string "hello" to get "olleh"

word = "hello"
result = []
for i in range(len(numbers)):
    result.append(word[-1-i])

print("".join(result))
```

```python
# Reverse the words in "hello world python" to get "python world hello"

words = "python world hello"
words = words.split()
result = []

for i in range(len(words)):
    result.append(words[-1-i])

print(" ".join(result))
    
    
```

```python
# Rotate [1, 2, 3, 4, 5] RIGHT by 2 positions to get [4, 5, 1, 2, 3]

numbers = [1, 2, 3, 4, 5]
k = 2

result = numbers[-k:] + numbers[:-k]
print(result)
```
