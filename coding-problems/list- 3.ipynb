# Finding Nth largest / Smallest pattern

```python
# Find the second largest number in [34, 12, 78, 23, 90, 45, 67]

numbers = [34, 12, 78, 23, 90, 45, 67]

if len(numbers) < 2:
    print(None)
    
else:
    largest = numbers[0]
    second_largest = numbers[1]

for number in numbers:
    if number > largest:
        second_largest = largest
        largest = number
    
    if number > second_largest and number != largest:
        second_largest = number
print(second_largest)
    
```

```python
# Find the second SMALLEST in [45, 12, 8, 67, 23, 5, 34]

numbers = [45, 12, 8, 67, 23, 5, 34]

if len(numbers) < 2:
    print(None)

else:
    if numbers[0] < numbers[1]:
        smallest = numbers[1]
        second_smallest = numbers[0]
    else:
        smallest = numbers[0]
        second_smallest = numbers[1]

for i in range(2,len(numbers)):
    number = numbers[i]

    if number < smallest:
        second_smallest = smallest
        smallest = number

    if number < second_smallest and number != smallest:
        second_smallest = number

print(second_smallest)
```

```python
# Find the THIRD largest in [34, 12, 78, 23, 90, 45, 67]

numbers = [34, 12, 78, 23, 90, 45, 67]

largest = float('-inf')
second_largest = float('-inf')
third_largest  = float('-inf')

for number in numbers:
    if number > largest:
        third_largest = second_largest
        second_largest = largest
        largest = number
    elif largest > number > second_largest:
        third_largest = second_largest
        second_largest = number
    elif second_largest > number > third_largest:
        third_largest = number


if third_largest == float('-inf'):
    print(None)
else:
    
    print(third_largest)
```

```python
# Find the MEDIAN (middle value) of [34, 12, 78, 23, 90]

numbers = [34, 12, 78, 23, 90]

numbers.sort()
n = len(numbers)

if n % 2 != 0:
    median = numbers[n//2]
    print(median)

else:
    median = (numbers[n//2 - 1] + numbers[n//2] ) / 2
    print(median)
```
